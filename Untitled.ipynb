# Kredi Kartı Dolandırıcılığı Tespiti için Sinir Ağı Modelleri

- Bu çalışmada, kredi kartı dolandırıcılığını tespit etmek amacıyla çeşitli sinir ağı modelleri geliştirilmiştir. Farklı model yapıları ve mimarileri kullanılarak, dolandırıcılık olaylarını doğru bir şekilde sınıflandırmak için en uygun yöntemler araştırılmıştır.

- İlk olarak, basit bir sinir ağı modeli oluşturulmuş, bu modelin temel özellikleri öğrenme süreci boyunca değerlendirilmiştir. Ardından, bir konvolüsyonel sinir ağı (CNN) modeli geliştirilmiş ve bu model, sıralı veri üzerinde özellik çıkarımı yaparak performansını artırmak hedeflenmiştir. Son olarak, derin bir sinir ağı (Deep NN) modeli kurulmuş, bu model daha fazla katman ve dropout teknikleri ile overfitting (aşırı uyum) sorununu minimize etmeyi amaçlamıştır.

- Her modelin performansı, eğitim ve doğrulama kayıpları ile doğruluk metrikleri kullanılarak detaylı bir şekilde analiz edilmiştir. Sonuçlar, modellerin genel performansını karşılaştırmak ve dolandırıcılık tespiti için en etkili yaklaşımı belirlemek amacıyla görselleştirilmiştir. ## Model kurmaya hazırlık.

```python
# bu kod veri setinizi eğitim ve test setlerine böler, böylece modelinizi eğitim verisi ile eğitebilir ve test verisi ile değerlendirebilirsiniz.

random_state = 123
features = [col for col in data.columns if col not in ['id', 'Class']]
X_train, X_test, y_train, y_test = train_test_split(data[features], data['Class'], test_size=0.3, random_state=random_state, stratify=data['Class'], shuffle=True)
```

```python
# bu kod parçası, eğitim ve test veri setlerini standartlaştırır.

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Bu fonksiyon, modelin test veri seti üzerindeki sonuçlarını değerlendirir ve bazı önemli performans metriklerini görselleştirir.

def show_model_results(model):
    pred = model.predict(X_test)
    pred = np.where(pred > 0.5, 1, 0)
    print(classification_report(y_test, pred))
    cm = confusion_matrix(y_test, pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()

def plot_history(history):
    training_loss = history.history['loss']
    validation_loss = history.history['val_loss']
    plt.plot(training_loss, label='Training Loss')
    plt.plot(validation_loss, label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()
```

```python
callbacks = [
    keras.callbacks.EarlyStopping(
        # `val_loss` iyileşmeyi bıraktığında eğitimi durdur
        monitor="val_loss",
        # "iyileşmeyi bıraktı" ifadesi "1e-2'den daha iyi olmama" olarak tanımlanır
        min_delta=1e-2,
        # "iyileşmeyi bıraktı" ifadesi "en az 2 epoch boyunca" olarak daha da tanımlanır
        patience=2,
        verbose=1,
    )
]
```

```python
print(y_train[:5])
```

## Basit bir derin öğrenme modeli

```python
def create_simple_nnmodel(neurons):
    model = Sequential()
    model.add(Dense(neurons, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model 
```

```python
start_time = time.time()
nn_model = create_simple_nnmodel(len(features))
history = nn_model.fit(X_train, y_train, epochs=20, batch_size=512, verbose=1, validation_split=0.2, callbacks=callbacks)
print(f'training time: {round(time.time()-start_time, 2)} seconds')
plot_history(history)
```

```python
nn_model.summary()
```

```python
show_model_results(nn_model)
```

## CNN Model

```python
def create_cnnmodel(input_length):
    # Define model architecture
    model = keras.Sequential([
        Conv1D(filters=5, kernel_size=10, activation="relu", input_shape=(input_length, 1), strides=2),
        MaxPooling1D(pool_size=2),
        Flatten(),
        Dense(32, activation="relu"),
        Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

start_time = time.time()
cnn_model = create_cnnmodel(len(features))
history = cnn_model.fit(X_train, y_train, epochs=20, batch_size=512, verbose=1, validation_split=0.2, callbacks=callbacks)
print(f'training time: {round(time.time()-start_time, 2)} seconds')
plot_history(history)
```

```python
show_model_results(cnn_model)
```

```python
cnn_model.summary()
```

## Deep neural network model

```python
def create_deep_nnmodel(neurons):
    model = Sequential()
    model.add(Dense(neurons, activation='relu'))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model 
```

```python
start_time = time.time()
deep_nn_model = create_deep_nnmodel(len(features))
history = deep_nn_model.fit(X_train, y_train, epochs=10, batch_size=512, verbose=1, validation_split=0.2, callbacks=callbacks)
print(f'training time: {round(time.time()-start_time, 2)} seconds')
plot_history(history)
```

```python
show_model_results(deep_nn_model)
```

```python
deep_nn_model.summary()
```

```python
# Eğitim sonuçlarını saklamak için bir liste oluşturun
results_list = []

# Basit Sinir Ağı Modeli
simple_nn_model = create_simple_nnmodel(neurons=10)
history_simple_nn = simple_nn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, verbose=0)
results_list.append(pd.DataFrame({
    'Model': 'Simple NN',
    'Epoch': range(1, len(history_simple_nn.history['loss']) + 1),
    'Loss': history_simple_nn.history['loss'],
    'Accuracy': history_simple_nn.history['accuracy']
}))

# CNN Modeli
cnn_model = create_cnnmodel(input_length=X_train.shape[1])
history_cnn = cnn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, verbose=0)
results_list.append(pd.DataFrame({
    'Model': 'CNN',
    'Epoch': range(1, len(history_cnn.history['loss']) + 1),
    'Loss': history_cnn.history['loss'],
    'Accuracy': history_cnn.history['accuracy']
}))

# Derin Sinir Ağı Modeli
deep_nn_model = create_deep_nnmodel(neurons=10)
history_deep_nn = deep_nn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, verbose=0)
results_list.append(pd.DataFrame({
    'Model': 'Deep NN',
    'Epoch': range(1, len(history_deep_nn.history['loss']) + 1),
    'Loss': history_deep_nn.history['loss'],
    'Accuracy': history_deep_nn.history['accuracy']
}))

# Listeyi tek bir veri çerçevesinde birleştirin
results = pd.concat(results_list, ignore_index=True)
```

```python
# Eğitim ve doğrulama kayıplarını görselleştirin
plt.figure(figsize=(14, 6))
sns.lineplot(data=results, x='Epoch', y='Loss', hue='Model', marker='o')
plt.title('Training and Validation Losses')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Eğitim ve doğrulama doğruluklarını görselleştirin
plt.figure(figsize=(14, 6))
sns.lineplot(data=results, x='Epoch', y='Accuracy', hue='Model', marker='o')
plt.title('Training and Validation Accuracies')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
